```python
from Bridge import Bridge
from Link import Link
from Source import Source
from Sink import Sink
from Vehicles import Truck 
from model import BangladeshModel, set_lat_lon_bound

df = pd.read_csv(path+'/data/simpleTransport.csv')
G= nx.DiGraph() #directed graph 
pos={}
sample = BangladeshModel()
agent_list=sample.schedule.agents 

for agent in agent_list:
    G.add_node(agent.name, label=agent.name) #adding them as nodes in the order of creation
    pos[agent.name]= agent.pos 
    print(agent.name)
#adding edges: 

for i in range(len(agent_list)):
    if i < len(agent_list)-1:
        G.add_edge(agent_list[i].name,agent_list[i+1].name)
        print('adding edge between'+ agent_list[i].name + 'and' + agent_list[i+1].name)

nx.draw(G,pos=pos, with_labels=True)

```
